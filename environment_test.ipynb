```python
from environment import OsuEnvironment
import os
import warnings
import logging
import time
from pynput.keyboard import Controller
if not os.path.exists('yolov5'):
    !git clone https://github.com/ultralytics/yolov5
    !pip install -r yolov5/requirements.txt

warnings.simplefilter("ignore", FutureWarning)
logging.getLogger('ultralytics').setLevel(logging.ERROR)
```

```python
# initialize the environment
env = OsuEnvironment()
keyboard = Controller()
```

```python
# run random songs
total_eppisode = 20
for episode in range(1, total_eppisode+1):
    state = env.reset()
    done = False
    count = 0
    rewards = 0
    env.pick_random_song()
    print(env.getSong())
    while not done and env.checking_connection():
        if env.song_begin():
            random_action = env.action_space.sample()
            state, reward, terminate, truncate = env.step(random_action)
            count += 1
            rewards += reward
            done = terminate or truncate

            if done:
                print(done)

        if env.lost_connection():
            break

    print(f"Episode: {episode} big update occur")
    print(f"Reward: {rewards}\n")
    env.return_to_song_selection_after_song()

    if episode == total_eppisode:
        keyboard.type("Finish training")
```
