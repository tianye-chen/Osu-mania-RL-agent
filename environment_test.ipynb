```python
# initialize the environment
env = OsuEnvironment()
keyboard = Controller()
```

```python
# run for  random songs
for episode in range(1, 50+1):
    state = env.reset()
    done = False
    count = 0
    rewards = 0
    env.pick_random_song()
    while not done and env.checking_connection():
        if env.song_begin():
            
            random_action = env.action_space.sample()
            state, reward, terminate, truncate = env.step(random_action)
            count += 1
            rewards += reward
            done = terminate or truncate

            if done:
                print(done)
        
            if env.getUpdateSignal() and not done:
                env.pause_game()
                print("small update occur")
                env.Updated()
                env.resume_game()

        if env.lost_connect():
            break

    print(f"Episode: {episode} big update occur")
    print(f"Reward: {rewards}")
    env.return_to_song_selection_after_song()
```
